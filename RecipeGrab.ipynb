```python
import pandas as pd 
import numpy as np 
from bs4 import BeautifulSoup as bs
import requests as r
import re
import json
```

```python
url = 'https://tasty.co/recipe/loaded-potato-chip-dip'
```

```python
html = r.get(url, timeout=1).content
soup = bs(html, 'lxml')

soup
```

```python
#INGREDIENTS
search = soup.find_all('ul', attrs= {'class':'list-unstyled xs-text-3'})

store = []
for i in search: 
    store.append(i.text)
    store1 = store[0].split('\n')

split = store[0].replace('\n','').split(' ')

store1 =[]
for i in split: 
    if i != '':
        store1.append(i)
        
ingredients = " ".join(store1)

ingredients
```

```python
#PREP

search = soup.find_all('ol', attrs= {'class':'prep-steps list-unstyled xs-text-3'})

store = []
for i in search: 
    store.append(i.text)
    store1 = store[0].split('\n')

split = store[0].replace('\n','').split(' ')

prep = ' '.join(split)
prep
```

```python
#TITLE

search = soup.find_all('h1', attrs= {'class':'recipe-name extra-bold xs-mb05 md-mb1'})

store = []
for i in search: 
    store.append(i.text)
    store1 = store[0].split('\n')

split = store[0].replace('\n','').split(' ')

prep = ' '.join(split)
prep
```

```python
urls = ['https://tasty.co/recipe/chili-beef-noodles',
'https://tasty.co/recipe/microwave-fried-rice',
'https://tasty.co/recipe/sweet-and-sour-chicken-noodles',
'https://tasty.co/recipe/sweet-and-sour-pork',
'https://tasty.co/recipe/garlic-sesame-salad-dressing',
'https://tasty.co/recipe/cauliflower-shrimp-fried-rice',
'https://tasty.co/recipe/cashew-chicken-stir-fry',
'https://tasty.co/recipe/veggie-garlic-noodles',
'https://tasty.co/recipe/orange-chicken-inspired-by-panda-express',
'https://tasty.co/recipe/lettuce-wraps']
```

```python
data = {}
data['Recipe'] = []

for i in urls:
    html = r.get(i, timeout=1).content
    soup = bs(html, 'lxml')
    search = soup.find_all('script', attrs= {'type':'application/ld+json'})
    data['Recipe'].append(search)
```

```python
#SAVE TO TXT

file1 = open("Recipe.txt","w") 
file1.write(str(data))
```

```python
#HTTP
urls = []

url = 'https://tasty.co/search?q=chinese+lunch'

html = r.get(url).content
soup = bs(html)
soup.prettify()

divTag = soup.find_all("div", {"class": "xs-block-grid-2 sm-block-grid-4"})

for tag in divTag:
    aTags = tag.find_all("a", {"class": "feed-item analyt-unit-tap"})
    for tag in aTags:
        urls.append(tag['href'])
```

```python
data = {}
data['Recipe'] = []

for i in urls:
    html = r.get(i, timeout=1).content
    soup = bs(html, 'lxml')
    search = soup.find_all('script', attrs= {'type':'application/ld+json'})
    data['Recipe'].append(search)
```

```python
#SAVE TO TXT

file1 = open("Recipe.txt","w") 
file1.write(str(data))
```
