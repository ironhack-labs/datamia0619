```python
def even(x):
    return(x % 2 == 0)
sample_space = [1, 2, 3, 4, 5, 6]
die_sides = len(sample_space)
even_roll = len([x for x in sample_space if even(x)])
even_probability = even_roll / die_sides
print(even_probability)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import binom
import numpy as np
n=10
p=0.4
binomial = binom(n,p)
print(binomial.pmf(3)) #3 is i
print(binomial.cdf(3))
print(binomial.mean())
print(binomial.var())
print(binomial.std())
print(binomial.rvs(10))
x = np.arange(0,10)
fig, ax = plt.subplots(1, 1)
ax.plot(x, binom.pmf(x, n, p), 'bo')
ax.vlines(x, 0, binom.pmf(x, n, p), colors='r', lw=5, alpha=0.8)
```

```python
import numpy as np
from scipy.stats import uniform
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1)
x = np.linspace(0,1, 100)
ax.plot(x, uniform.pdf(x))
```

```python
fig, ax = plt.subplots(1, 1)
a=1
b=3
x = np.linspace(a,b, 100)
y = uniform.pdf(x, a, b)
ax.plot(x, y)
```

```python
import math
from scipy.stats import norm
normal = norm(0, math.sqrt(9)) 
print(normal.pdf(4))
print(normal.cdf(2))       
print(normal.rvs()) 
fig, ax = plt.subplots(1, 1)
x = np.linspace(-3,3,1000)
y = norm.pdf(x)
ax.plot(x,y)
plt.show()
```
