####  What is the average Purchase Price?

```python
# your answer here
orders.amount_spent.min()
```

#### Select all the customers we have in Spain

```python
# your answer here
set(orders[orders.Country == "Spain"]['CustomerID'])
```

#### How many customers do we have in Spain?
##### Hint : Use value_counts()

```python
# your answer here
len(orders[orders.Country == "Spain"]['CustomerID'].value_counts())
```

#### Select all the customers who have bought more than 50 items ?

```python
subset = orders[['CustomerID', 'Quantity']].groupby('CustomerID').count()
subset[subset.Quantity > 50]
```

#### Select orders from Spain that are above 50 items

```python
# your answer here
orders[((orders.Country == "Spain") & (orders.Quantity > 50))]
```

#### Select all free orders

```python
# your answer here
orders[orders.amount_spent == 0]
```

#### Select all orders that are 'lunch bag'
#### Hint: Use string functions

```python
# your answer here
orders[orders.Description.str.startswith("lunch bag")]
```

#### Select all orders that are made in 2011 and are 'lunch bag'  

```python
# your answer here
orders[((orders.Description.str.startswith("lunch bag")) & (orders.year == 2011))]
```

#### Show the frequency distribution of the amount spent in Spain.

```python
# your answer here
%matplotlib inline
subset = orders[orders.Country == "Spain"].groupby('amount_spent').count()['InvoiceNo']
subset.hist(bins=200)
```

#### Select all orders made in the month of August

```python
# your answer here
orders[orders.month == 8]
```

#### Select how many orders are made by countries in the month of August
##### Hint: Use value_counts()

```python
# your answer here
orders[orders.month == 8].groupby('Country').Country.value_counts()
```

#### What's the  average amount of money spent by country

```python
# your answer here
orders.groupby('Country').amount_spent.mean()
```

#### What's the most expensive item?

```python
# your answer here
orders[orders.UnitPrice == orders.UnitPrice.max()]['Description']
```

#### What was the average amount spent per year ?
