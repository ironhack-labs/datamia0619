#### Group the employees based on their number of years of employment. What are the average salaries in each group?

```python
# your answer here
groups = employee.groupby('Years')

for year in employee['Years'].unique():
    x = 's' if year > 1 else 's'
    salary = round(groups.get_group(year)['Salary'].mean())
    print("Average annual salary following {} year{} is ${} 000".format(year, x, salary))
```

####  What is the average Salary per title?

```python
# your answer here
for title in employee['Title'].unique():
    emp = round(employee[employee['Title'] == title]['Salary'].mean())
    vowel = ('e', 'a', 'i', 'o', 'u', 'un')
    x = 'an' if title.lower().startswith(vowel) else 'a'
    print("Average annual salary for {} {} is ${} 000".format(x, title, emp))
```
