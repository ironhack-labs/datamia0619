## Import the librarys

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Import the dataset

```python
data = pd.read_csv('train.csv')
data.head()
```

## List the top three things that most houses dont have

```python
null_data = data.isnull().sum(axis = 0)
SalePrice_Stats = data['SalePrice'].describe()

print("Top 3 things that houses don't have\n")
print(null_data.sort_values(ascending=False)[:3])
```

## Displaying some information about the Sales Price

```python
print("Some indepth information about house sales prices\n")
print(SalePrice_Stats)
```

## The top things that impact purchasing a house(Main Puropse)

```python
print("Biggest impacts on buying a house\n")
print(data.corr()['SalePrice'].sort_values(ascending=False)[1:5])
```

## Creating a chart to break down the data, and display it.

```python
print("Full pie chart of what is direct outcome of if house is purchased.")
labels = ["SalePrice","OverallQual","GrLivArea","GarageCars","GarageArea","TotalBsmtSF",
          "1stFlrSF","FullBath","TotRmsAbvGrd","YearBuilt","YearRemodAdd","GarageYrBlt",
          "MasVnrArea ","Fireplaces","BsmtFinSF1","LotFrontage","WoodDeckSF","2ndFlrSF",
          "OpenPorchSF","HalfBath","LotArea","BsmtFullBath","BsmtUnfSF","BedroomAbvGr",
          "ScreenPorch","PoolArea","MoSold","3SsnPorch","BsmtFinSF2","BsmtHalfBath",
          "MiscVal","Id","LowQualFinSF","YrSold","OverallCond","MSSubClass","EnclosedPorch","KitchenAbvGr"]
sizes = [x for x in data.corr()['SalePrice'].sort_values(ascending=False)]

explode = (0, 0.1, 0)
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')
plt.show()
```

## Sort/Simplify the data to present to someone looking on the market

```python
print("Simplified pie chart of what is direct outcome of if house is purchased.")
labels = ["OverallQual","GrLivArea","GarageCars","GarageArea"]
sizes = [x for x in data.corr()['SalePrice'].sort_values(ascending=False)[1:5]]

explode = (0, 0.1, 0)
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')
plt.show()
```
