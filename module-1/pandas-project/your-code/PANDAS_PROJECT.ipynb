```python
#Converting Fatal Column to 0 and 1


f = df['Fatal (Y/N)']
count3 = 0

for i in f: 
    try:
        if i == 'UNKNOWN': 
            f[count3] = float('nan')
        elif 'n' in i.lower() or i == 'F':
            f[count3] = 0 
        elif i == 'Y':
            f[count3] = 1 
        else: 
            f[count3] = float('nan')
    except: 
        pass 
    count3 += 1
```

```python
#Converting female 0 and male 1
s = df['Sex']
count4 = 0

for i in s: 
    try: 
        if 'm' in i.lower(): 
            s[count4] = 1
        elif 'f' in i.lower(): 
            s[count4] = 0  
        else: 
            s[count4] = float('nan')
    except: 
        pass 
    count4 += 1
    
```
