```python
import pandas as pd

# Import comma-separated variable file
data = pd.read_csv('GSAF5.csv', encoding="ISO-8859-1")
data.head()
```

```python
data.shape
```

```python
# Count null entries in each column
null_cols = data.isnull().sum()
null_cols[null_cols > 0]
```

```python
# 1. Remove all columns with > 90% of null entries
drop_cols = list(null_cols[null_cols > (data.shape[0]*0.9)].index)
data = data.drop(drop_cols, axis=1)
data.head()
```

```python
#2. Update column names, so they don't contain extra whitespace
print(data.columns)
data = data.rename(columns={'Sex ':'Sex', 'Species ':'Species'})
data.columns
```

```python
# Find columns with low variance
i = 0
for col in data:
    if len(set(data[col])) < 10:
        print(data.columns[i])
        print(set(data[col]))
    i = i + 1
```

```python
# 3. Clean-up values for 'Sex' so its either "M", "F", or "O" for Other
data['Sex'] = data['Sex'].str.replace('M ', 'M')
data['Sex'] = data['Sex'].str.replace('lli', 'O')
data['Sex'] = data['Sex'].str.replace('N', 'O')
data['Sex'] = data['Sex'].str.replace('.', 'O')
data['Sex'] = data['Sex'].fillna('O')
print(set(data['Sex']))
```

```python
# 4. Clean-up values for 'Fatal (Y/N)' so its either "Y", "N", or "U" for Unknown
data['Fatal (Y/N)'] = data['Fatal (Y/N)'].str.replace('#VALUE!', 'U')
data['Fatal (Y/N)'] = data['Fatal (Y/N)'].str.replace('N ', 'N')
data['Fatal (Y/N)'] = data['Fatal (Y/N)'].str.replace(' N', 'N')
data['Fatal (Y/N)'] = data['Fatal (Y/N)'].str.replace('F', 'Y')
data['Fatal (Y/N)'] = data['Fatal (Y/N)'].str.replace('UNKNOWN', 'U')
data['Fatal (Y/N)'] = data['Fatal (Y/N)'].str.replace('n', 'N')
data['Fatal (Y/N)'] = data['Fatal (Y/N)'].fillna('U')
print(set(data['Fatal (Y/N)']))
```

```python
# 5. Clean-up values for 'Type' so its either:
# 'Boat', 'Unprovoked', 'Sea Disaster', 'Invalid', 'Provoked'
data['Type'] = data['Type'].str.replace('Boating', 'Boat')
print(set(data['Type']))
```

```python
data.dtypes
```

```python
# 6. Convert 'Age' column to int, and drop any records that are NOT numbers
before = len(data)

for index, row in data.iterrows():
    try:
        int(row['Age'])
    except:
        data.drop(index, inplace=True)
data['Age'] = data['Age'].astype('int')
data['Age'].dtype

after = len(data)
print('Number of records dropped: ', str(before - after))
```

```python
data.shape
```

```python
# Find # of rows where case number is the same as BOTH Case Number.1 and Case Number.2
i = 0
for index, row in data.iterrows():
    if row['Case Number'] == row['Case Number.1'] and row['Case Number'] == row['Case Number.2']:
        i = i + 1
print(i)
```

```python
# 7. Drop Case Number.1 and Case Number.2 rows since it is ~99% duplicate data
data = data.drop(columns=['Case Number.1','Case Number.2'])
```

```python
# Find # of rows where href formula is the same as href
i = 0
for index, row in data.iterrows():
    if row['href'] == row['href formula']:
        i = i + 1
print(i)
```

```python
# 8. Drop href formula as its ~99% duplicate data
data = data.drop(columns=['href formula'])
data.head()
```
