```python
import json
from pandas.io.json import json_normalize
import requests
import pandas as pd

# Determine number of records at the Harvard Art Museum
url = 'https://api.harvardartmuseums.org/object?apikey=b15b3500-99ac-11e9-a66a-a9decfd62676'
response = requests.get(url)
results = response.json()
num_records = results['info']['totalrecords']
num_pages = results['info']['pages']
num_pages
```

```python
# Get all records in the Harvard Art Museum
# WARNING: This will run for several hours
i = 1
all_records = []
response = requests.get(url + '&size=100&page=1')
results = response.json()
df = pd.DataFrame(results['records'])
while(i < num_pages):
    i += 1
    response = requests.get(url + '&size=100&page=' + str(i))
    new_df = pd.DataFrame(response.json()['records'])
    df = pd.concat([df,new_df], sort=True)
    if (i %10 == 0): print(df.shape)
df.head()
```

```python
# Create a fun subset of the data
cols = ['Object', 'Century', 'Department', 'Acquired']
data = df[['title', 'century','department', 'accessionyear']]
data.columns = cols
data.shape
```

```python
data2 = data.drop_duplicates()
```

```python
data3 = data2.sort_values(by=['Acquired'])
data3.head(100)
```

```python
# Export to CSV
export_file = data3.to_csv('hartvard.csv', index=False)
```
