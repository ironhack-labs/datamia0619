```python
def amazon_reviews(start=0,Page_N = 1):
    df = pd.DataFrame() 
    try:
        for i in range(start, Page_N):
            #Pause
            sleep(10)
            #Get
            #url = 'https://www.amazon.com/National-Geographic-Magazine/product-reviews/B008E88Q74/ref=cm_cr_arp_d_viewopt_srt?filterByStar=critical&pageNumber='+(str(Page_N))+'&sortBy=recent'
            url = 'https://www.amazon.com/National-Geographic-Magazine/product-reviews/B008E88Q74/ref=cm_cr_arp_d_viewopt_srt?filterByStar=critical&pageNumber='+(str(i+1))+'&sortBy=recent'
            cookies = dict(cookies_are='working')
            request = r.get(url, cookies=cookies, timeout=5)
            html = request.content 
            #Search
            soup = bs(html, 'lxml')
            soup_search = soup.find_all('span',{'class':""})
            search = re.findall('<span class="">[A-Z].{1,190}</span>, <span class="">[A-Z].{1,800}</span> ?',str(soup_search))
            #Clean
            lstrip = [i.lstrip('<span class="">') for i in search]
            rstrip = [i.rstrip('</span>') for i in lstrip]
            split = [i.split('</span>, <span class="">') for i in rstrip]
            #Catgorize
            Titles = [i[0] for i in split] 
            Bodies = [i[1] for i in split] 
            #Dataframe 
            df_temp = pd.DataFrame()
            df_temp['Title'] = Titles
            df_temp['Body'] = Bodies
            df = df.append(df_temp, ignore_index = True)
            print(len(df))
        
    except: 
        pass
    
    return df
```
