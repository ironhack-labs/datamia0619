```python
import numpy as np
import pandas as pd
from scipy.spatial.distance import pdist, squareform
```

```python
books = pd.read_csv('books.csv')
ratings =pd.read_csv('ratings.csv')
```

```python
ratings.head(5)
```

```python
books.head()
```

```python
booktitle=books[['book_id','title']]
```

```python
booktitle.head()
```

```python
bookrate=pd.merge(ratings, booktitle, on='book_id')
```

```python
bookrate.head()
```

```python
bookagg = bookrate.groupby(['user_id', 'book_id','title']).agg({'rating':'mean'}).reset_index()
```

```python
bookagg
```

```python
bookrate_pivot = bookagg.pivot_table(values='rating', 
                                                columns='user_id', 
                                                index='title', 
                                                aggfunc='mean').fillna(0)
```

```python
bookrate_pivot
```

```python
book_dist = pd.DataFrame(1/(1 + squareform(pdist(bookrate_pivot, 'euclidean'))),
                         index=bookrate_pivot.index, columns=bookrate_pivot.index)

book_dist.head()
```

```python
allusers={}
users = list(bookrate['user_id'].unique())

for user in users[0:3]:
    datax=bookrate[(bookrate['user_id']==user) & (bookrate["rating"]>0)]
    nestedRecs = []
    for book in bookrate['title']:
        nestedRecs.append(book_dist[book].sort_values(ascending=False)[1:6].index)
    lst = []
    for item in nestedRecs:
        for i in item:
            if i not in datax['title']:
                lst.append(i)
    import pandas as pd
    top5=pd.Series(lst).value_counts().head(5).index
    allusers[user]=top5
```
