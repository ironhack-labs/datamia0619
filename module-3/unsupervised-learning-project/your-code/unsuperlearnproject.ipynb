```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
iris_scale=scaler.fit_transform(X)
```

```python
iris_scale
```

```python
from sklearn.cluster import KMeans
```

```python
XX = iris_scale
kmeans4 = KMeans(n_clusters=4).fit(XX)
```

```python
X['clusters'] = kmeans4.labels_
X['clusters'].value_counts()
```

```python
from sklearn.metrics import silhouette_score
y_pred = kmeans4.predict(XX)
score = silhouette_score(XX, y_pred, metric='euclidean')
score
```

```python
XX = iris_scale
kmeans2 = KMeans(n_clusters=2).fit(XX)
X['clusters2'] = kmeans2.labels_
print(X['clusters2'].value_counts())
y_pred2 = kmeans2.predict(XX)
score2 = silhouette_score(XX, y_pred2, metric='euclidean')
score2
```

```python
XX = iris_scale
kmeans3 = KMeans(n_clusters=3).fit(XX)
X['clusters3'] = kmeans3.labels_
print(X['clusters3'].value_counts())
y_pred3 = kmeans3.predict(XX)
score3 = silhouette_score(XX, y_pred3, metric='euclidean')
score3
```

```python
from yellowbrick.cluster import KElbowVisualizer

model = KMeans()
visualizer = KElbowVisualizer(model, k=(1,6))
visualizer.fit(XX)
visualizer.poof
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 3))
for k in range(1,5):
    kmeans = KMeans(n_clusters=k)
    t = kmeans.fit_predict(XX)
    plt.subplot(1,6,k)
    plt.scatter(XX[:, 0], XX[:, 1], c=kmeans.labels_, cmap='viridis')
    plt.xlabel('k='+str(k))
plt.tight_layout()
plt.show()
```

```python
#both the elbow visualizer and gap statistics seem to indicate 3 clusters make the most sense
```

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=0.5, min_samples=5).fit(XX)
dbscan.labels_
X['labels_DBSCAN']=dbscan.labels_
```

```python
dblabel=DBSCAN(eps=0.5, min_samples=5).fit_predict(XX)
```

```python
X['labels_DBSCAN'].value_counts()
#three clusters 
```

```python
scoredb = silhouette_score(XX, X['labels_DBSCAN'], metric='euclidean')
scoredb
```

```python
from sklearn.decomposition import PCA

pca = PCA()
pca.fit(XX)
pcaX = pca.transform(XX)

fig = plt.figure(figsize=(6, 6))

plt.scatter(pcaX[:, 0], pcaX[:, 1], c = X['clusters3'], cmap='viridis')
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
```

```python
from sklearn.manifold import TSNE

fig, ax = plt.subplots(1, 4, figsize=(15, 7))
for i, perp in enumerate([5, 30, 50, 100]):
    tsne = TSNE(perplexity=perp)
    x_embedded = tsne.fit_transform(XX)
    ax[i].scatter(x_embedded[:, 0], x_embedded[:, 1], c=X['clusters3'],cmap='viridis')
    ax[i].set_title("Perplexity = {}".format(perp))
```
