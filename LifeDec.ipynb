```python
import pandas as pd 
import numpy as np 
```

```python
State = ['CA', 'DC', 'NY', 'VA', 'WA', 'TX', 'MA','IL','ML','PA','NC','GA','CO','NJ','FL']
Av_Salary = [90.5, 73, 71.5, 71, 117.3, 68, 70.5, 65.3 , 67.3, 64, 67.3,65, 67, 67.3, 67.3]
population = [39.5, .633, 8.6, 8.5, 2.6, 28.7,6.9,12.7, 6.04, 12.8,10.4,10.5, 5.6,8.9,21.3]
Jobs = [3521, 1683, 1322, 1243, 1075, 877, 824, 659, 509, 455, 428, 321, 313, 313, 302]
CPI = [151.7,158.4,139.1,100.7, 110.7, 91.5, 131.6, 94.5, 129.7, 101.7, 94.9, 89.2, 105.6, 125.1, 97.9]
Weather = [10, 7.5, 1, 7, 1, 10, 3, 2, 1, 5, 9, 9, 7, 2, 9]
People = [5, 5, 2, 4, 7, 8, 9, 1, 5, 3, 8, 7.5, 8, 1, 4]
```

```python
df = pd.DataFrame(index=State)
```

```python
df.head()
```

```python
df['JobsPerCapita'] = np.array(Jobs)/np.array(population)
df['Weather'] = Weather
df['People'] = People
df['S/CPI'] = np.array(Av_Salary)/np.array(CPI)
df
```

```python
def standardize(array): 
    mean = np.mean(array)
    stdev = np.std(array)
    new = []
    
    for i in array:
        x = (i-mean)/stdev
        new.append(x)
    
    return np.array(new)

def min_max(array):
    mini = np.min(array)
    maxi = np.max(array)
    new = []
    
    for i in array:
        x = (i-mini)/(maxi-mini)
        new.append(x)
    
    return np.array(new)
```

```python
for i in df.columns: 
    df[i] = min_max(df[i])
    
df
```

```python
df['Weightless'] = df.JobsPerCapita + df.Weather + df.People + df['S/CPI']
```

```python
df.sort_values('Weightless', ascending=False)
```

```python
w_df = pd.DataFrame(index = State)
```

```python
def weighting(options, weather, people, ratio): 
    w_df['JobsPerCapita'] = df.JobsPerCapita*options
    w_df['Weather'] = df.Weather*weather
    w_df['People'] = df.People*people
    w_df['Iratio'] = df['S/CPI']*ratio
    w_df['Weight'] = w_df.JobsPerCapita + w_df.Weather + w_df.People + w_df.Iratio
    
    return w_df.sort_values('Weight', ascending=False)
    
        
```
